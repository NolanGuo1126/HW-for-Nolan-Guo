1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

1. Accuracy
Scenario: Customer feedback surveys for a product.
Rationale: In this context, accuracy measures the overall percentage of true responses (satisfied and unsatisfied customers) relative to total responses. High accuracy indicates that the survey effectively captures customer sentiment, which is important for business assessments. However, it’s crucial to ensure that this measure doesn't overshadow reporting of false positives or negatives, particularly if customer satisfaction is a deciding factor for future product development or marketing strategies.

2. Sensitivity
Scenario: HIV testing in a high-risk population.
Rationale: Sensitivity is crucial in medical scenarios where failing to detect an actual positive case could have life-threatening outcomes. A highly sensitive test ensures that most individuals who are HIV-positive are correctly identified, allowing for timely treatment and reducing the risk of transmission. This metric is essential in public health to control and manage disease spread effectively.

3. Specificity
Scenario: Fraud detection systems in banking.
Rationale: Specificity is vital for minimizing disruptions caused by false alarms when assessing transactions. A system with high specificity accurately identifies legitimate transactions, reducing the number of false positives (transactions flagged as fraudulent when they are not). This efficiency is critical for maintaining customer trust and ensuring smooth banking operations without unnecessary scrutiny.

4. Precision
Scenario: Cancer detection algorithms in radiology.
Rationale: Precision is key when interpreting results from imaging technologies (e.g., MRIs or CT scans) that suggest the presence of cancer. A high precision rate means that a significant percentage of the positive identifications of cancer are actual cases. This is important to avoid unnecessary biopsies and anxiety for patients due to false positives, thereby improving patient management and care outcomes.

Summary of Distinctions
Accuracy provides a broad measure of the model's overall correctness but can obscure issues related to specific types of errors.
Sensitivity is critical for identifying true positive instances, vital in healthcare contexts where missing a condition can have severe implications.
Specificity focuses on correctly identifying negatives, which is important in processes like fraud detection where false positives can create operational challenges.
Precision emphasizes the correctness of positive predictions, ensuring that interventions based on identified positives are justified, particularly in medical diagnostics.
By understanding these nuances, organizations can select and use the appropriate metrics based on their specific goals and the consequences of their decisions.

````python
import pandas as pd
import numpy as np

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"
ab = pd.read_csv(url, encoding="ISO-8859-1")

# Initial EDA: Show the first few rows and basic info
print(ab.head())
print(ab.info())

# Step 2: Pre-processing as per requirements
# Remove specified columns
ab_reduced = ab.drop(columns=['Weight_oz', 'Width', 'Height'])

# Drop all remaining rows with NaN entries
ab_reduced_noNaN = ab_reduced.dropna()

# Set data types
ab_reduced_noNaN['Pub year'] = ab_reduced_noNaN['Pub year'].astype(int)
ab_reduced_noNaN['NumPages'] = ab_reduced_noNaN['NumPages'].astype(int)
ab_reduced_noNaN['Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Display data types of the DataFrame
print(ab_reduced_noNaN.dtypes)
````

````python
from sklearn.model_selection import train_test_split

# Set random seed for reproducibility
seed = 42

# Create the training and testing sets (80/20 split)
ab_reduced_noNaN_train, ab_reduced_noNaN_test = train_test_split(ab_reduced_noNaN, test_size=0.2, random_state=seed)

# Reporting the number of observations in training and testing sets
print(f'Training observations: {len(ab_reduced_noNaN_train)}')
print(f'Testing observations: {len(ab_reduced_noNaN_test)}')
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Prepare the data for classification
y = pd.get_dummies(ab_reduced_noNaN_train["Hard_or_Paper"])['H']  # Convert target to binary
X = ab_reduced_noNaN_train[['List Price']]  # Feature for prediction

# Initialize and fit the Decision Tree Classifier
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(X, y)

# Visualizing the decision tree
tree.plot_tree(clf)
````

> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.
